### 【總體四】平日/促銷日AOV差異

將促銷日訂為 11/23 當日，其他日期皆為平日，觀察兩者的 AOV 是否呈現差異。

#### [進階] 依現有欄位新增欄位

判斷式：
    
    if condition1:
      result1
    elif condition2:
      result2
    elif condition3:
      result3
    ...
    else:
      result4

```python
# [進階] 根據現有欄位的資料新增欄位，判斷那天是平日或促銷日
def label_onsale(row):
   if row['datetime']=='2018-11-23':
      return 1
   else:
      return 0

df['blackfri'] = df.apply(label_onsale, axis=1)  # axis=1 意思是一次處理一筆 row 資料
df
```

【練習題】 請依序執行上方程式碼，並計算平日/促銷日的平均客單價有沒有差異？(欄位：blackfri, value [USD])

```python
df.groupby('blackfri')['value [USD]'].mean()
```

## 電商資料產生自訂欄位的方法

新增 order_id 欄位：
* 把 index 的數字加上 14783 之後設為一個新欄位，並將欄位名稱取名為 order_id2。
* 14783 只是一個自定義的數字，由於這份資料僅是該公司訂單資料的部份擷取，因此我將這份資料的範圍假設為第14783~39781號的訂單資料(不重複流水號)。

```python
df['order_id2'] = df.index + 14783
df
```

新增 prod_cnt 欄位：
* 新增一個命名為 rand_prod_cnt 的欄位，將該欄位內隨機填上介於 a~b 之間的數值。
* 使用隨機填值，需要用到 numpy 套件，語法為：np.random.randint(a, b+1, df.shape[0])
* 由於是隨機填值，因此每次長出來的數字都不一樣。

```python
import numpy as np
```

```python
df['rand_prod_cnt'] = np.random.randint(1, 6, df.shape[0])
df
```

## 小組練習

```python
df['datetime'] = pd.to_datetime(df['booking_date'])
df['datetime'] = pd.to_datetime(df['datetime']).dt.year
df.groupby(['datetime','country_destination'])['price'].mean()
```

```python
def country_tw_or_not(row):
    if row['country_destination'] == 'TW':
        return 'TW'
    else:
        return 'Not TW'
df['country_tw_or_not'] = df.apply(country_tw_or_not, axis=1)
```

```python
df['datetime'] = pd.to_datetime(df['booking_date'])
df['datetime'] = pd.to_datetime(df['datetime']).dt.year
df.groupby(['datetime','country_tw_or_not'])['price'].mean()
```
